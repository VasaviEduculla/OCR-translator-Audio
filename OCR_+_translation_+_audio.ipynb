```
import wget
```

```
url='https://i.stack.imgur.com/mk1jX.jpg'
```

```
wget.download(url)
```

```
reader=easyocr.Reader(['ta'])
translator=Translator()
```

```
import PIL
from PIL import ImageDraw
image=PIL.Image.open("mk1jX.jpg")
```

```
image
```

```
#txt="https://hindi.rapidleaks.com/wp-content/uploads/2018/10/sant-kabir-das-dhoe-1280x720.jpg"
```

```
bounds=reader.readtext("mk1jX.jpg",add_margin=0.55,width_ths=0.7,link_threshold=0.8,decoder='beamsearch',blocklist='=.')
bounds
```

```
def draw_boxes(image,bounds,color='yellow',width=2):
  draw=ImageDraw.Draw(image)
  for bound in bounds:
    p0,p1,p2,p3=bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0],fill=color,width=width)
  return image
```

```
draw_boxes(image,bounds)
```

```
text_list=reader.readtext("mk1jX.jpg",add_margin=0.55,width_ths=0.7,link_threshold=0.8,decoder='beamsearch',blocklist='=.',paragraph=True,detail=0)
text_list
```

```
text_list=text_list[0]
```

```
text_list
```

```
print(translator.detect(text_list))   #shows the language and confidence level
```

```
txt = translator.translate(text_list,src='ta')  # english, indian slang it gets converted , ta -> tamil
print(txt.text)
```

```
trans_txt = gTTS(txt.text) #default US accent it gets converted to
trans_txt.save("US_eng.mp3")
```

```
Audio("US_eng.mp3",autoplay=True)
```

```
trans_txt = gTTS(txt.text,lang='hi') 
trans_txt.save("Indian_eng.mp3")
```

```
Audio("Indian_eng.mp3",autoplay=True)
```
